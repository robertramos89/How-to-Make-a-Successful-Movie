```python
engine = create_engine(connection_str)
if database_exists(connection_str) == False: create_database(connection_str)
else: print('The database already exists.')
```

```python
# import basics dataframe from part 1
df_title = pd.read_csv('Data/title_basics.csv.gz', low_memory=False)
df_title.head()
```

```python
# import ratings dataframe from part 1
df_ratings = pd.read_csv('Data/title_ratings.csv.gz', low_memory=False)
df_ratings.head()
```

```python
# Add a new column into the titles dataframe with genre split out
df_title['genre_split'] = df_title['genres'].str.split(',')
df_title
```

```python
# Explode the list of genres into new columns
exploded_genres = df_title.explode('genre_split')
exploded_genres
```

```python
# Save unique genre's into variable
unique_genres = sorted(exploded_genres['genre_split'].unique())
unique_genres
```

```python
# save tconst and split genre columns into a new dataframe
title_genres = exploded_genres[['tconst', 'genre_split']].copy()
title_genres.head()
```

```python
#make a genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
#make new integer genre_id and drop string genres
title_genres['genre_id'] = title_genres['genre_split'].map(genre_map)
title_genres = title_genres.drop(columns='genre_split')
title_genres.head()
```

```python
genre_lookup = pd.DataFrame({'Genre_Name': genre_map.keys(), 'Genre_ID': genre_map.values()})
genre_lookup.head()
```

Export Title Genres Table to MySQL

```python
## Calculate max string lengths for object columns
key_len = title_genres['tconst'].fillna('').map(len).max()

## Create a schema dictonary using Sqlalchemy datatype objects
title_genre_schema = {
    "tconst": String(key_len+1), 
    'genre_id':Integer()}
```

```python
# Save to sql with dtype and index=False
title_genres.to_sql('title_genres',engine,dtype=title_genre_schema,if_exists='replace',index=False)
```

```python
q = """
SELECT *
FROM title_genres
LIMIT 5
"""
pd.read_sql(q, engine)
```
