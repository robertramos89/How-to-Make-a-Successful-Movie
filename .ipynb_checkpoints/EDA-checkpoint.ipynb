# EDA

```python
#import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
# Load dataframe for year 2000
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
# Load dataframe for year 2001
df_2001 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001.head()
```

```python
# Drop the first NaN row in the dataframe for the year 2000
df_2000 = df_2000.iloc[1:,:]
df_2000
```

```python
# Drop the first NaN row in the dataframe for the year 2001
df_2001 = df_2001.iloc[1:,:]
df_2001
```

```python
#Concatenate the previous results and new results.
merged = pd.concat([df_2000, df_2001], ignore_index=True)

display(merged.head(2),merged.tail(2))
```

```python
# save the merged dataframe into a csv
merged.to_csv('Data/tmdb_combined_results.csv.gz',compression='gzip',index=False)
```

```python
# Shorten dataframe too only relevant columns needed for analysis
df = merged[['imdb_id','budget','revenue','certification']]
df
```

```python
# Filter out dataframe withoout 0's for budget and revenue
df_filtered = df[(df['budget']>0) & (df['revenue']>0)]
df_filtered.head()
```

# Questions

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Create barplot
plt.figure(figsize=(10,8))
ax= sns.countplot(x = 'certification', data = df_filtered)
ax.set_title('Number of Movies Per Certification Category')
plt.xlabel('Certifications')
plt.ylabel('# of Movies')


# barplot annotation
for bar in ax.patches:
    ax.annotate(format(bar.get_height(), '0.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```

What is the average revenue per certification category?

```python
#import tool to annotate y axis with commas code from link below
#https://stackoverflow.com/questions/64895794/matplotlib-how-do-i-add-commas-to-marker-labels

import matplotlib.ticker as ticker
```

```python
plt.figure(figsize = (10,8))
ax = sns.barplot(x = 'certification', y = 'revenue', data = df_filtered, ci=None)
ax.set_title('Average Revenue per Certification')
plt.xlabel('Certifications')
plt.ylabel('Avg Revenue ($)')
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda y, p: format(int(y),','))
)

# barplot annotation
for bar in ax.patches:
    ax.annotate('{:,.2f}'.format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```

What is the average budget per certification category?

```python
plt.figure(figsize = (10,10))
ax = sns.barplot(x = 'certification', y = 'budget', data = df_filtered, ci=None)
plt.xlabel('Certification')
plt.ylabel('Average Budget ($)')
ax.set_title('Certifications')
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda y, p: format(int(y),','))
)
#barplot annotation
for bar in ax.patches:
    ax.annotate('{:,.2f}'.format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```
